# Random Forest

An ensemble of decision trees used for classification and regression tasks.

- Typically trained using the "bagging" method:
    - Trained on randomly selected subsets of the training data.
- The algorithm constructs multiple decision trees and combines them to produce a more accurate and stable prediction, introducing additional randomness while growing the trees.
- Final Decision:
    - Soft voting is used for classification.
    - Averaging is used for regression.

![](pict/random_forest.jpeg)

## Random Forest Classifier

```python
from sklearn.datasets import load_breast_cancer
X, y = load_breast_cancer(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

### n_estimators
This parameter is used to specify the number of decision trees to be included in the model.

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10, random_state=0)
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

### Feature Importances
The *feature_importances_* attribute returns the significance of each feature in building the model using the Random Forest algorithm.
- The sum of the importances for all features equals 1.

```python
rf.feature_importances_
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,4))
plt.bar(load_breast_cancer().feature_names, rf.feature_importances_)
plt.xticks(rotation=90);
```

## Random Forest Regressor

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

### n_estimators

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=10, random_state=0)
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```
